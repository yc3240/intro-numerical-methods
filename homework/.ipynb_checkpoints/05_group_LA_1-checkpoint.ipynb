## Question 1 - P-Norms

The $p$-norm of a vector is defined as

$$
    ||v||_p = \left[ \sum^N_{i=1} (v_i)^p \right ]^{1/p}.
$$

**(a)** Write a program that calculates the p-norm of a vector $\vec{x} \in \mathbb{R}^n$.  Note that $p \notin \mathbb Z^+$ but $p > 0$.  You can use `numpy.linalg.norm` to check your work.

```python
def p_norm(v, p):
    if p == numpy.inf:
        return numpy.max(v)
    else:
        return numpy.sum(v**p)**(1.0/p)

v = numpy.random.rand(10)
p = numpy.inf
print v
print numpy.linalg.norm(v, p)
print p_norm(v,p)
```

**(b)** A *unit p-norm* implies for cases where $p$ is finite that

$$
    1 = ||v||_p = \left( \sum_{i=1}^2 |v_i|^p \right)^{\frac{1}{p}}
$$

where here we are limiting ourselves to $v \in \mathbb R^2$.  For $p = \infty$ we have

$$ 
    1 = \max_{1 \leq i \leq 2} |v_i|
$$ 

Find explicit formulations for the vectors $\vec{v} \in \mathbb{R}^2$ with a unit p-norm. 

By definition given,

$$\left( v_1^p + v_2^p \right)^{\frac{1}{p}} =1$$

Therefore,

$$v_1^p + v_2^p =1$$

**(c)** With this formula, plot various unit p-norms.  As $p \rightarrow \infty$ what do you observe?

```python
fig = plt.figure()
fig.set_figwidth(fig.get_figwidth()*1.2)
axes = fig.add_subplot(1, 1, 1)
vx = numpy.linspace(0.0,1.0,100)

for n in xrange(1,5):
    p = 2.0**n
    vy = (1.0 - vx**p)**(1.0/p)    
    #plot mirror contour along vx = 0 -> 1 -> 0 -> -1 -> 0
    #when vx goes 1 -> 0 or -1 -> 0
    #vx need flipping
    x = numpy.concatenate((vx, vx[::-1], -vx, -vx[::-1])) 
    y = numpy.concatenate((vy,-vy[::-1], -vy,  vy[::-1]))
    axes.plot(x,y,'--',label = "p=%s"%(p))
axes.axis('equal')
axes.legend(loc=7)
axes.set_title("Unit Vector Sets")
axes.set_xlabel("x")
axes.set_ylabel("y")
axes.set_xlim(-1.2, 1.2)
axes.set_ylim(-1.2, 1.2)
plt.show()
```

## Question 2

Given a $A \in \mathbb R^{2 \times 2}$, plot the unit vectors $\hat{i}$ and $\hat{j}$ and the unit circle in your favorite p-norm and the transformed unit vectors and circle (should be an ellipse).

```python
A = numpy.random.rand(2,2)
print "Matrix A = "
print A

p = 5.0
vx = numpy.linspace(0.0,1.0,100)
vy = (1.0 - vx**p)**(1.0/p)
x = numpy.concatenate((vx, vx[::-1], -vx, -vx[::-1])) 
y = numpy.concatenate((vy,-vy[::-1], -vy,  vy[::-1]))
unit_circle = numpy.array([x,y])
unit_i = numpy.array([1.0, 0.0])
unit_j = numpy.array([0.0, 1.0])

#Plot p-norm
fig = plt.figure()
fig.set_figwidth(fig.get_figwidth()*2.0)
axes = []
axes.append(fig.add_subplot(1, 2, 1))
axes[0].plot(unit_circle[0,:],unit_circle[1,:],label = "Unit circle of norm %s"%(p))
axes[0].arrow(0.0,0.0,1.0,0.0,head_width=0.05,head_length=0.05,label = 'i')
axes[0].arrow(0.0,0.0,0.0,1.0,head_width=0.05,head_length=0.05,label = 'j')
axes[0].axis('equal')
axes[0].set_title('Norm of order %s' %(p))
axes[0].set_xlabel('x')
axes[0].set_ylabel('y')
axes[0].set_xlim(-1.5, 1.5)
axes[0].set_ylim(-1.5, 1.5)
#Plot transformed p-norm
axes.append(fig.add_subplot(1, 2, 2))
trans_circle = numpy.dot(A, unit_circle)
axes[1].plot(trans_circle[0,:],trans_circle[1,:],label = "Unit circle of norm %s"%(p))
[xi, yi] = numpy.dot(A,numpy.array([1.0, 0.0]))
[xj, yj] = numpy.dot(A,numpy.array([0.0, 1.0]))
axes[1].arrow(0.0,0.0,xi,yi,head_width=0.05,head_length=0.05,label = 'i')
axes[1].arrow(0.0,0.0,xj,yj,head_width=0.05,head_length=0.05,label = 'j')
axes[1].plot()
axes[1].axis('equal')
axes[1].set_title("Norm Transformed by matrix A")
axes[1].set_xlabel('x')
axes[1].set_ylabel('y')
axes[1].set_xlim(-1.5, 1.5)
axes[1].set_ylim(-1.5, 1.5)
plt.show()
```
